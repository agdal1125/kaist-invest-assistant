```python
import os
import json
import pandas as pd

NH_HOME = os.getenv("NH_HOME")

etf_dataset_path = f"{NH_HOME}/invest_agent/workflow/etf/etf_items.csv"
etf_dataset = pd.read_csv(etf_dataset_path, sep="|")
```

```python
etf_dataset
```

```python
df = etf_dataset[["ETF_명", "구성종목명", "구성비율(%)", "ETF_종목코드"]]

etf_dict = {}
total_etf_name = {}

for _, row in df.iterrows():
    etf_code = row["ETF_종목코드"].strip()
    etf_name = row["ETF_명"].strip()
    stock_name = str(row["구성종목명"]).strip()
    ratio = float(row["구성비율(%)"])
    
    total_etf_name[etf_name] = [etf_name]
    
    if etf_name not in etf_dict:
        etf_dict[etf_name] = {
			"etf_code": etf_code,
			"ratio": {}
		}
    etf_dict[etf_name]["ratio"][stock_name] = ratio
```

```python
etf_dict
```

```python
# save etf_dict
save_path = "etf_dict.json"
with open(save_path, "w") as f:
    json.dump(etf_dict, f, indent=4, ensure_ascii=False)
    
save_path = "etf_name_database.json"
with open(save_path, "w") as f:
    json.dump(total_etf_name, f, indent=4, ensure_ascii=False)
```

```python
# 종목명으로 구성된 list를 stock db 형태로 저장
stoc_items_raw = list(df["구성종목명"].unique())
stoc_items_db = {}
for item in stoc_items_raw:
    item = item.strip()
    stoc_items_db[item] = [item]

# save stoc_items_db
with open("etf_database.json", "w") as f:
    json.dump(stoc_items_db, f, indent=4, ensure_ascii=False)
```
