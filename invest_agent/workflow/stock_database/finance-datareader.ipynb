```python
# stocks = fdr.StockListing('S&P500') # S&P500: 503 종목  
stocks = fdr.StockListing('NASDAQ') # 나스닥 (NASDAQ): 4천+ 종목
# stocks = fdr.StockListing('NYSE') # 뉴욕증권거래소 (NYSE): 3천+ 종목
```

```python
stocks["Symbol"], stocks["Name"]
```

```python
stock_data = []
for symbol, name in zip(stocks["Symbol"], stocks["Name"]):
    stock_data.append({
		"symbol": symbol,
		"name": name,
	})
    
with open("stock_data.json", "w") as f:
    json.dump(stock_data, f, indent=4)
```
